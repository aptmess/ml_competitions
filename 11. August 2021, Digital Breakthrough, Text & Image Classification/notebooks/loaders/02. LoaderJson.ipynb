```python
def generate_api(typology, s, l):
    API = """https://opendata.mkrf.ru/v2/museum-exhibits/$?f={"data.typology.name":{"$search":"""
    typ = lambda typology: f""""{typology}"""
    s2 = """"}}"""
    s3 = lambda s, l: f"&s={s}&l={l}"
    return API + typ(typology) + s2 + s3(s, l)
```

```python
API_TOKEN = 'fa8e356bb2858559f4d18da092d9fba7c934f9197240d984305edaaf67999baf'


def get_data(category,
             start=0, 
             end=100000, 
             step=1000, 
             output_path=DATA_PATH, 
             filename='loaded_full_data.csv'):

    result_arr = []
    missed = -1
    print(f"Loading {category}")

    for st in tqdm(np.arange(start, end+step, step), 
                   total=int(end / step), 
                   desc='Loading..'):
        try:
            response = requests.get(generate_api(category, st, step), 
                                    headers={'X-API-KEY': API_TOKEN})
            response.raise_for_status()
            
        except Exception as ex:
            continue
        response_data = response.json()
        full_response_data = response_data.get('data')
        if full_response_data is not None:
            for item_data in full_response_data:

                guid = missed
                description = missed
                typology = missed
                url = missed
                name = missed
                tech_technologies = missed


                guid = item_data.get('_id', missed)
                data = item_data.get('data', missed)
                
                
                if data is not missed:
                    images = data.get('images', missed)


                    if images is not '':
                        if isinstance(images, list):
                            list_url = images[0]
                            url = list_url.get('url', missed)
                    
                    typ = data.get('typology', missed)
                    if typ is not missed:
                        typology = typ.get('name', missed)


                    tech = data.get('technologies', missed)
                    if tech is not missed:
                        if isinstance(tech, list):
                            tech = [t for t in tech if isinstance(t, str)]
                            tech_technologies = (
                                ' '.join(tech) if tech is not [] else missed
                                )


                    description = data.get('description', missed)
                    name = data.get('name', missed)

                item = {
                    'guid': guid,
                    'description': description,
                    'typology': typology,
                    'url': url,
                    'name': name,
                    'technologies': tech_technologies
                }
                result_arr.append(item)
            
            idx = len(result_arr)
        if idx % 10000 == 0:
            sub = pd.DataFrame(result_arr)
            sub.to_csv(str(output_path / filename), index=False)
            print(f'saved: {idx} rows')
            # with open(str(output_path / filename), 'w', encoding='utf8', ensure_ascii=False) as file:
            #     json.dump(result_arr, file, indent=4)
    sub = pd.DataFrame(result_arr)
    sub.to_csv(str(output_path / filename), index=False)
    print(f'saved: {idx} rows')
    # with open(str(output_path / filename), 'w', encoding='utf8') as file:
    #     json.dump(result_arr, file, indent=4)
```
