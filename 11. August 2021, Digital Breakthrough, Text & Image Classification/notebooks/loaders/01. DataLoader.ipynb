## Описание соревнования

### Условия

Обучение искусственного интеллекта на распознавание вида или категории предмета музейного фонда при внесении нового предмета в каталог.

Основным источником данных является ФГИС «государственного каталога Музейного фонда Российской Федерации», Сбор данных происходит в автоматическом режиме посредством внутренних информационных систем Минкульутуры России.

Информация в датасете представлена с 2011 года по настоящее время.

### Подробное описание датасета.

#### Задача

Необходимо по описанию или изображению определить категорию предмета музейного фонда.

Формат данных в `train.csv` и `test.csv`:

- `guid` - идентификатор предмета 
- `description` - описание предмета
- `typology` - типология 

Дополнительная обучающая выборка - `train_url_only.csv`.

Дополнительная выборка характеризуется значительно большим объемом данных по количеству предметов, однако файлы изображений по ним отсутствуют, но их можно получить по ссылкам, хранящимся в столбце `url`. 

Столбцы файла дополнительной выборки:

- `guid` - идентификатор предмета (но файл изображения с таким именем отсутствует)

- `description` - описание предмета
- `typology` - типология 
- `url` - ссылка на файл изображения в открытом датасете

Тестовая выборка

Для некоторых предметов в тестовом наборе могут быть только изображения, для других - только текстовые описания, для третьих - и то и другое. 
Имена файлов  в наборе изображений совпадают с значением столбца `guid`.

### Метрика

F1-score

### Дополнительные источники 

- Разрешается использовать любые открытые данные о музейном фонде для расширения и обогащения обучающей выборки

- Запрещается напрямую сопоставлять представленную в задаче тестовую выборку с открытыми данными и использовать таким образом готовые правильные ответы
